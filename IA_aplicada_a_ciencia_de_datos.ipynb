### **2da Pregunta:** ¿Cuál fue la facturación por categoria em 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# prompt: Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.
# En "df", tenemos las siguientes columnas:
# categoria: categorías de los productos
# precio_unitario: precio de una unidad del producto
# cantidad: cantidad vendida del producto
# envio: costo de envío total de la compra.
# Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.
# Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.

# Calculate revenue for each row (price * quantity + shipping)
df_combinado['facturacion_item'] = df_combinado['precio_unitario'] * df_combinado['cantidad'] + df_combinado['envio']

# Group by category and sum the calculated facturacion_item
facturacion_por_categoria = df_combinado.groupby('categoria')['facturacion_item'].sum().reset_index()

# Sort the categories by revenue for better visualization
facturacion_por_categoria = facturacion_por_categoria.sort_values(by='facturacion_item', ascending=False)

# Create a horizontal bar chart
plt.figure(figsize=(10, 8))
sns.barplot(x='facturacion_item', y='categoria', data=facturacion_por_categoria, palette='viridis')
plt.title('Facturación por Categoría en 2023')
plt.xlabel('Facturación Total (MXN$)')
plt.ylabel('Categoría')
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
````

* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

````python
# prompt: Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras apiladas que muestre las ventas por trimestre en relación con el método de pago de Zoop.
# En "df", tenemos las siguientes columnas:
# fecha: con las fechas de las compras en formato datetime (aaaa-mm-dd)
# facturacion: con los valores de las ventas
# metodo_pago: con el método de pago elegido en la compra
# Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.
# Crea el código.
# Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización. cuando sea posible, Usa el idioma español en los comentarios.

# Asegurarse de que la columna 'fecha' sea de tipo datetime
df_combinado['fecha'] = pd.to_datetime(df_combinado['fecha'])

# Asegurarse de que la columna 'facturacion' exista (si no, calcularla)
if 'facturacion' not in df_combinado.columns:
    df_combinado['facturacion'] = df_combinado['precio_unitario'] * df_combinado['cantidad'] + df_combinado['envio']

# Crear una columna para el trimestre
df_combinado['trimestre'] = df_combinado['fecha'].dt.to_period('Q')

# Agrupar por trimestre y método de pago, sumando la facturación
facturacion_trimestral_metodo = df_combinado.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().reset_index()

# Convertir la columna 'trimestre' a string para el eje x del gráfico
facturacion_trimestral_metodo['trimestre'] = facturacion_trimestral_metodo['trimestre'].astype(str)

# Crear el gráfico de barras apiladas
plt.figure(figsize=(10, 7))
sns.barplot(x='trimestre', y='facturacion', hue='metodo_pago', data=facturacion_trimestral_metodo, palette='viridis')
plt.title('Facturación por Trimestre y Método de Pago en 2023')
plt.xlabel('Trimestre')
plt.ylabel('Facturación Total (MXN$)')
plt.legend(title='Método de Pago')
plt.tight_layout() # Ajustar el diseño
plt.show()
````

````python
# prompt: guardar las visualizaciones como imagen. Coloca cada uno de los gráficos que generamos dentro de una función y quiero que se aplique el siguiente procedimiento a cada una de ellas:
# nombre_imagen = nombre_funcion()
# nombre_imagen.savefig('nombre_imagen.png', bbox_inches='tight')
# nombre_imagen.close()
# Las funciones son:
# metodo_pago()
# ventas_mensuales()
# facturacion_por_categoria()
# ventas_por_trimestre()
# proporcion_cashback()
# distribucion_calificaciones()
# box_plot_edad_sexo()
# Reemplaza la variable "nombre_imagen" por "grafico_" seguido del nombre de la función que genera el gráfico.
# Crea el código
# Utiliza el lenguaje Python para crear el código y no mostrar las imágenes.

# Redefine the functions to generate the plots and return the figure object

def metodo_pago():
    """Generates and returns the payment method bar chart figure."""
    metodos_de_pago = df_combinado['metodo_pago'].value_counts().reset_index()
    metodos_de_pago.columns = ['Metodo de Pago', 'Cantidad']

    total_first_two_methods = metodos_de_pago.loc[0:1, 'Cantidad'].sum()
    total_all_methods = metodos_de_pago['Cantidad'].sum()
    porcentaje = (total_first_two_methods / total_all_methods) * 100

    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    bars = ax.bar(metodos_de_pago['Metodo de Pago'], metodos_de_pago['Cantidad'], color=ROJO_1)

    ax.set_title("Métodos de pago más utilizados en 2023", fontsize=18, color=GRIS_5)
    ax.set_xlabel("Método de Pago", fontsize=12, color=AZUL_1)
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.set_ylabel("")
    ax.set_yticks([])

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)

    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 0), ha='center', va='bottom', fontsize=12, color=AZUL_1)

    text_to_add = f"$\\bf{porcentaje:.2f}$% de los clientes utilizan Tarjeta de Crédito o Transferencia \npara el pago. Sería interesante considerar asociaciones con \nbancos para la creación de un Zoop Pay con el fin de fidelizar \na nuestra clientela."
    ax.text(1.5, 2500, text_to_add, fontsize=10, color=AZUL_3, ha='left', va='top')

    plt.tight_layout()
    return fig

def ventas_mensuales():
    """Generates and returns the monthly sales line plot figure."""
    df_combinado['fecha'] = pd.to_datetime(df_combinado['fecha'])
    df_combinado['facturacion'] = df_combinado['precio_unitario'] * df_combinado['cantidad'] + df_combinado['envio']

    meses = {
        'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
        'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
        'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
    }

    df_combinado['mes'] = df_combinado['fecha'].dt.strftime('%B')
    ventas_mensuales_data = df_combinado.groupby('mes')['facturacion'].sum().reindex(meses.keys())

    fig, ax = plt.subplots(figsize=(12, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    ax.plot(ventas_mensuales_data.index, ventas_mensuales_data, marker='o', color=AZUL_1, linewidth=2)

    ax.set_title("Ventas totales mensuales en 2023", fontsize=18, color=GRIS_5)

    def format_million(x, pos):
        return f'MXN$ {x/1000000:.2f} M'

    formatter = plt.FuncFormatter(format_million)
    ax.yaxis.set_major_formatter(formatter)
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    month_labels = [meses[month] for month in ventas_mensuales_data.index]
    ax.set_xticks(ventas_mensuales_data.index)
    ax.set_xticklabels(month_labels)
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)

    ax.set_xlabel("")
    ax.set_ylabel("")

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.grid(axis='y', color=GRIS_2, linestyle='-', linewidth=0.5)

    highlight_indices = [4, 7, 10, 11]
    circle_radius = 200000

    for i, idx in enumerate(highlight_indices):
        x_coord = ventas_mensuales_data.index[idx]
        y_coord = ventas_mensuales_data.iloc[idx]
        circle = Circle((idx, y_coord), circle_radius, color=ROJO_1, fill=False, linewidth=2)
        ax.add_patch(circle)
        text_x = idx - (circle_radius * 0.005)
        text_y = y_coord
        ax.text(text_x, text_y, str(i + 1), color=ROJO_1, fontsize=14, ha='right', va='center', weight='bold')


    plt.tight_layout()
    return fig


def facturacion_por_categoria():
    """Generates and returns the revenue by category bar chart figure."""
    df_combinado['facturacion'] = df_combinado['precio_unitario'] * df_combinado['cantidad'] + df_combinado['envio']
    facturacion_categoria_data = df_combinado.groupby('categoria')['facturacion'].sum().reset_index()
    facturacion_categoria_data = facturacion_categoria_data.sort_values(by='facturacion', ascending=False)

    first_bar_value = facturacion_categoria_data.iloc[0]['facturacion']
    total_revenue_all_categories = facturacion_categoria_data['facturacion'].sum()
    porcentaje = (first_bar_value / total_revenue_all_categories) * 100

    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    bars = ax.barh(facturacion_categoria_data['categoria'], facturacion_categoria_data['facturacion'], color=AQUA_1)

    ax.set_title("Facturación por categoría de producto en 2023", fontsize=18, color=GRIS_5)
    ax.set_xlabel("")
    ax.set_xticks([])
    ax.set_ylabel("")
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    for bar in bars:
        xval = bar.get_width()
        yval = bar.get_y() + bar.get_height()/2
        ax.text(xval, yval, f'{xval:,.0f}', ha='left', va='center', fontsize=12, color=AZUL_1, weight='bold')

    text_to_add = f"$\\bf{porcentaje:.2f}$% de las ventas corresponden únicamente a las ventas de Electrónica.\nEsto equivale a casi el doble en relación con todas las demás categorías\npresentes en el portafolio de Zoop."
    ax.text(0.4e7, 5, text_to_add, fontsize=12, color=AZUL_3, ha='left', va='center')

    plt.tight_layout()
    return fig


def ventas_por_trimestre():
    """Generates and returns the stacked bar chart for sales by quarter and payment method figure."""
    df_combinado['fecha'] = pd.to_datetime(df_combinado['fecha'])
    if 'facturacion' not in df_combinado.columns:
        df_combinado['facturacion'] = df_combinado['precio_unitario'] * df_combinado['cantidad'] + df_combinado['envio']

    df_combinado['trimestre'] = df_combinado['fecha'].dt.to_period('Q')
    ventas_trimestre_data = df_combinado.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack()
    ventas_trimestre_data.index = ventas_trimestre_data.index.astype(str)

    custom_palette = [BLANCO, GRIS_3, GRIS_1, GRIS_3]

    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    ventas_trimestre_data.plot(kind='bar', stacked=True, ax=ax, color=custom_palette)

    ax.set_title("Ventas por Trimestre y Método de Pago en 2023", fontsize=18, color=GRIS_5)
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    plt.xticks(rotation=0)

    def format_million_no_decimal(x, pos):
        return f'MXN$ {x/1000000:.0f} M'

    formatter = plt.FuncFormatter(format_million_no_decimal)
    ax.yaxis.set_major_formatter(formatter)
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    ax.set_xlabel("")
    ax.set_ylabel("")

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.grid(axis='y', linestyle=':', color=GRIS_2, linewidth=0.7)

    legend_labels = ventas_trimestre_data.columns.tolist()
    legend_handles = [plt.Rectangle((0,0),1,1, color=custom_palette[i]) for i in range(len(custom_palette))]
    ax.legend(legend_handles, legend_labels, title='Método de Pago por trimestre de 2023', loc='upper left', bbox_to_anchor=(1, 1), facecolor=GRIS_1, edgecolor=GRIS_5)

    plt.tight_layout(rect=[0, 0, 0.85, 1])
    return fig


def proporcion_cashback():
    """Generates and returns the donut chart for cashback proportion figure."""
    cashback_proporcion_data = df_combinado['cashback'].value_counts()
    colors = [AZUL_1 if label == 'Sí' else GRIS_3 for label in cashback_proporcion_data.index]

    fig, ax = plt.subplots(figsize=(6, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    wedges, texts, autotexts = ax.pie(cashback_proporcion_data,
                                      labels=cashback_proporcion_data.index,
                                      autopct='%1.1f%%',
                                      colors=colors,
                                      startangle=180,
                                      explode=(0, 0),
                                      wedgeprops={'width': 0.4})

    for text, autotext, color in zip(texts, autotexts, colors):
        text.set_color(color)
        text.set_fontsize(12)
        autotext.set_color(BLANCO)
        autotext.set_fontsize(12)

    for autotext in autotexts:
        autotext.set_position((autotext.get_position()[0] * 0.8, autotext.get_position()[1] * 0.8))

    ax.set_title("¿Participan en el programa de cashback?", fontsize=18, color=GRIS_5)
    ax.axis('equal')

    plt.tight_layout()
    return fig


def distribucion_calificaciones():
    """Generates and returns the distribution of ratings histogram figure."""
    mean_nota = df_combinado['nota'].mean().round(2)
    most_common_nota = df_combinado['nota'].value_counts().idxmax()
    most_common_frequency = df_combinado['nota'].value_counts().max()
    annotation_text = f"Promedio de calificaciones = {mean_nota}\n Calificación más común = {most_common_nota}\n Frecuencia más común = {most_common_frequency}"

    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    sns.histplot(data=df_combinado, x='nota', bins=11, kde=False, color=ROJO_1, ax=ax)

    ax.set_title("Distribución de las evaluaciones de las compras en 2023", fontsize=18, color=GRIS_5)
    ax.set_xlabel('Calificación (0-10)', fontsize=12, color=AZUL_1)
    ax.set_ylabel('Frecuencia', fontsize=12, color=AZUL_1)
    ax.set_xticks(range(0, 11))
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_axisbelow(True)
    ax.grid(axis='y', color=GRIS_2, linestyle='-', linewidth=0.5)

    ax.annotate(annotation_text,
                xy=(8.7, 2250),
                xytext=(2, 1800),
                fontsize=12,
                color=AZUL_1,
                ha='left', va='top',
                bbox=dict(boxstyle='round,pad=0.5', fc=BLANCO, ec=GRIS_3, linewidth=1.5),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2', color=AZUL_1, linewidth=1.5))

    plt.tight_layout()
    return fig

def box_plot_edad_sexo():
    """Generates and returns the boxplot for age distribution by biological sex figure."""
    stats_by_sex = df_combinado.groupby("sexo_biologico")["edad"].agg(['min', lambda x: x.quantile(0.25), 'median', lambda x: x.quantile(0.75), 'max'])
    stats_by_sex.columns = ['Mínimo', '1º quartil', '2º quartil', '3º quartil', 'Máximo']

    sex_labels = df_combinado['sexo_biologico'].unique().tolist()
    index_for_first_sex = stats_by_sex.index.get_loc(sex_labels[0])
    index_for_second_sex = stats_by_sex.index.get_loc(sex_labels[1])

    stats_sex_0_boxplot = stats_by_sex.iloc[index_for_first_sex]
    annotation_text_0 = f"Mínimo = {stats_sex_0_boxplot['Mínimo']}\n 1º quartil = {stats_sex_0_boxplot['1º quartil']}\n 2º quartil = {stats_sex_0_boxplot['2º quartil']}\n 3º quartil = {stats_sex_0_boxplot['3º quartil']}\n Máximo = {stats_sex_0_boxplot['Máximo']}"

    stats_sex_1_boxplot = stats_by_sex.iloc[index_for_second_sex]
    annotation_text_1 = f"Mínimo = {stats_sex_1_boxplot['Mínimo']}\n 1º quartil = {stats_sex_1_boxplot['1º quartil']}\n 2º quartil = {stats_sex_1_boxplot['2º quartil']}\n 3º quartil = {stats_sex_1_boxplot['3º quartil']}\n Máximo = {stats_sex_1_boxplot['Máximo']}"


    fig, ax = plt.subplots(figsize=(10, 6), facecolor=GRIS_1)
    fig.patch.set_facecolor(GRIS_1)
    ax.set_facecolor(GRIS_1)

    boxplot_palette = [AQUA_1, AZUL_1]

    sns.boxplot(x='sexo_biologico', y='edad', data=df_combinado, palette=boxplot_palette, ax=ax)

    ax.set_title("Distribución de la Edad por Sexo Biológico de los clientes", fontsize=18, color=GRIS_5)
    ax.set_xlabel("")
    ax.set_ylabel('Edad', fontsize=12, color=AZUL_1)
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_axisbelow(True)
    ax.grid(axis='y', color=GRIS_2, linestyle='-', linewidth=0.5)

    boxplot_x_order = [t.get_text() for t in ax.get_xticklabels()]

    if sex_labels[0] == boxplot_x_order[0]:
        annotation_text_for_first_box = annotation_text_0
    else:
        annotation_text_for_first_box = annotation_text_1

    ax.annotate(annotation_text_for_first_box,
                xy=(-0.1, 45),
                xytext=(-0.4, 50),
                fontsize=10,
                color=AZUL_1,
                ha='left', va='top',
                bbox=dict(boxstyle='round,pad=0.5', fc=BLANCO, ec=GRIS_3, linewidth=1.5),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2', color=AZUL_1, linewidth=1.5))

    if sex_labels[0] == boxplot_x_order[1]:
        annotation_text_for_second_box = annotation_text_0
    else:
         annotation_text_for_second_box = annotation_text_1

    ax.annotate(annotation_text_for_second_box,
                xy=(0.9, 45),
                xytext=(0.6, 50),
                fontsize=10,
                color=AZUL_1,
                ha='left', va='top',
                bbox=dict(boxstyle='round,pad=0.5', fc=BLANCO, ec=GRIS_3, linewidth=1.5),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2', color=AZUL_1, linewidth=1.5))


    plt.tight_layout()
    return fig

# Export the figures as images
grafico_metodo_pago = metodo_pago()
grafico_metodo_pago.savefig('grafico_metodo_pago.png', bbox_inches='tight')
plt.close(grafico_metodo_pago) # Use plt.close() with the figure object

grafico_ventas_mensuales = ventas_mensuales()
grafico_ventas_mensuales.savefig('grafico_ventas_mensuales.png', bbox_inches='tight')
plt.close(grafico_ventas_mensuales)

grafico_facturacion_por_categoria = facturacion_por_categoria()
grafico_facturacion_por_categoria.savefig('grafico_facturacion_por_categoria.png', bbox_inches='tight')
plt.close(grafico_facturacion_por_categoria)

grafico_ventas_por_trimestre = ventas_por_trimestre()
grafico_ventas_por_trimestre.savefig('grafico_ventas_por_trimestre.png', bbox_inches='tight')
plt.close(grafico_ventas_por_trimestre)

grafico_proporcion_cashback = proporcion_cashback()
grafico_proporcion_cashback.savefig('grafico_proporcion_cashback.png', bbox_inches='tight')
plt.close(grafico_proporcion_cashback)

grafico_distribucion_calificaciones = distribucion_calificaciones()
grafico_distribucion_calificaciones.savefig('grafico_distribucion_calificaciones.png', bbox_inches='tight')
plt.close(grafico_distribucion_calificaciones)

grafico_box_plot_edad_sexo = box_plot_edad_sexo()
grafico_box_plot_edad_sexo.savefig('grafico_box_plot_edad_sexo.png', bbox_inches='tight')
plt.close(grafico_box_plot_edad_sexo)
````
